```python
import json
import matplotlib.pyplot as plt
import plotly.express as px

# Load JSON data into a list of dictionaries
def load_jsonl(file_path):
    with open(file_path, 'r') as file:
        return [json.loads(line) for line in file]

#houses_data = load_jsonl("/content/houses.jsonl")
# employee_data = load_jsonl("/content/employees_germans_1.json")
review_data = load_json("/content/reviews_germans_1.json")
def load_json(path):
    with open(path, "r") as f:
      return json.loads(f.read())


# Filter out houses with prices in billions
# houses_data = [house for house in houses_data if 'price' in house
#               and isinstance(house['price'], (int, float)) and
#               house['price'] < 1000000000]/*

# Function to format price labels with "million" suffix
#def format_price_label(price):
#    if price < 1000000:
#        return str(price)
 #   return f'{price / 1000000:.1f} million'

# Function to plot a scatter plot for prices vs a numeric column
def plot_scatter(data, column):
    if column != 'review_score':  # Exclude the "price" column
        plt.figure(figsize=(10, 6))
        prices = [house['review_score'] for house in data]
        values = [house[column] for house in data]
        plt.scatter(values, prices, s=10, alpha=0.5)
        # Use smaller dots (s=10) and make them slightly transparent (alpha=0.5)
        plt.title(f'Scatter Plot of review_score vs {column}')
        plt.xlabel(column)
        plt.ylabel('review_score')
        plt.grid(True)  # Add a grid

        # Format the y-axis labels with "million" suffix
        # plt.gca().get_yaxis().set_major_formatter(
        #    plt.FuncFormatter(lambda x, loc: format_price_label(x)))

        plt.show()

import json
import matplotlib.pyplot as plt


def load_jsonl(file_path):
    with open(file_path, 'r') as file:
        return [json.loads(line) for line in file]

houses_data = load_jsonl("C:\\Users\\Niko\\Desktop\\Handin2\\data\\houses.jsonl")

for x in houses_data:
    if "year" not in x:
        houses_data.remove(x)

def head(data, n=5):
    return data[:n]

# Usage
print(head(houses_data))

def info(data):
    if not data:
        return "Empty Dataset"
    sample_row = data[0]
    print("Number of rows:", len(data))
    print("Columns and Types:")
    for key in sample_row:
        print(f" - {key}: {type(sample_row[key])}")

# Usage
info(houses_data)


def count_nulls(data):
    null_counts = {key: 0 for key in data[0].keys()}
    for row in data:
        for key in row:
            if row[key] is None:
                null_counts[key] += 1
    return null_counts

# Usage
print(count_nulls(houses_data))



def count_duplicates(data):
    seen = set()
    duplicates = 0
    for row in data:
        row_tuple = tuple(sorted(row.items()))
        if row_tuple in seen:
            duplicates += 1
        else:
            seen.add(row_tuple)
    return duplicates

# Usage
print(count_duplicates(houses_data))

def describe(data, column):
    values = [row[column] for row in data if isinstance(row[column], (int, float))]
    mean = sum(values) / len(values)
    min_value = min(values)
    max_value = max(values)
    return {
        "count": len(values),
        "mean": mean,
        "min": min_value,
        "max": max_value
    }

# Usage for a numeric column, e.g., 'price'
print(describe(houses_data, 'price'))


def value_counts(data, column):
    counts = {}
    for row in data:
        key = row[column]
        if key in counts:
            counts[key] += 1
        else:
            counts[key] = 1
    return counts

# Usage
for column in houses_data[0]:
    print(f"Value counts for {column}:")
    print(value_counts(houses_data, column))
    print()



# Function to plot a bar chart for a categorical column
def plot_bar_chart(data, column):
    counts = value_counts(data, column)
    plt.figure(figsize=(10, 6))
    plt.bar(counts.keys(), counts.values())
    plt.title(f'Bar Chart of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()

# Plot bar charts for each categorical column
for column in houses_data[0]:
    if not isinstance(houses_data[0][column], (int, float)):  # Categorical columns
        plot_bar_chart(houses_data, column)

# Plot scatter plots for prices vs each numeric column (excluding "price")
#for column in review_data[1]:
    #if isinstance(review_data[1][column], (String)) and column != 'review_score':
      # Numeric columns (excluding "price")
        #plot_scatter(review_data, 'worker_id')
reviews =  int(review_data[1]['review_score'])

fig_yp = px.scatter(x=reviews, y ='worker_id', title="")
fig_yp.show()

```
